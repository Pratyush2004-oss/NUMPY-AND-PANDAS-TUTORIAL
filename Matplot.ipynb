```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,200)
x
```

```python
y = np.sin(x)
y
```

```python
graph1 = plt.plot(y)
graph1
```

```python
plt.scatter(x,y)
plt.xlabel('x axis')
plt.ylabel('Y axis')
plt.title('This is my graph')
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
x,y
```

```python
plt.scatter(x,y)
```

```python
colors = np.random.rand(50)
sizes = np.random.rand(50)*100
plt.scatter(x,y,c=colors,s=sizes,alpha=0.5)
```

```python
a = ['a','b','c','d','e']
y = np.random.rand(5)
plt.barh(a,y)
y
```

```python
a = ['a','b','c','d','e']
y = np.random.rand(5)
y
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y,c='#808080',marker='P')
```

```python
x = [5,7,8,9,14]
y = [5,7,8,9,14]
plt.scatter(x,y,marker='x')
```

```python
x = [5,7,8,9,14]
y = [5,7,8,9,14]
colors = ['red','yellow','blue','green','orange']
plt.scatter(x,y, c = colors)
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x,y,z)
plt.show()
```
